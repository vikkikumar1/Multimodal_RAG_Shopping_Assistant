# Data Exploration for Multimodal RAG

This notebook loads the product data and performs exploratory analysis to understand the catalog.

```python
import sys
sys.path.append("..")  # to import from project modules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from utils.data_loader import load_product_data
from utils.config import RAW_DATA_DIR
from pathlib import Path

# Set style
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# Load data
df = load_product_data()
print(f"Loaded {len(df)} products")
df.head()
```

```python
# Basic information
df.info()
```

```python
# Check missing values
missing = df.isnull().sum()
missing[missing > 0]
```

### Column Statistics

```python
# Numerical stats
df.describe()
```

### Categorical Distributions

```python
# Categories
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
cols = ['masterCategory', 'gender', 'season', 'usage']
for ax, col in zip(axes.flatten(), cols):
    if col in df.columns:
        df[col].value_counts().plot(kind='bar', ax=ax, title=col)
        ax.tick_params(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Price distribution (if price column exists)
if 'price' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df['price'], bins=30, kde=True)
    plt.title("Price Distribution")
    plt.xlabel("Price ($)")
    plt.show()
```

```python
### Image Availability
```

```python
# Count products with local images
image_dir = RAW_DATA_DIR / "images"
if image_dir.exists():
    df['has_image'] = df['image_path'].apply(lambda x: bool(x) and Path(x).exists())
    image_count = df['has_image'].sum()
    print(f"Products with images: {image_count} / {len(df)} ({100*image_count/len(df):.1f}%)")
else:
    print("Image directory not found.")
```
